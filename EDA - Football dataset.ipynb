# Converting columns to the required type

```python
data1.Wage 
```

##### Inference:

The data in the wage column have euro sign and are followed by 'K' which denote thousand or 'M' which denote a million. The euro sign have to be removed and 'M' and 'K' have to be replaced with 0's so analysis can be done on the column.

This is the same case with the Value and release clause columns

```python
def curtonum (x): # function to remove euro sign and add zeros appropirately according to the suffix
    x = x.strip()
    temp =x[1:-1]
    if x[-1].upper() == 'M':
        temp=float(temp)*1000000
    elif x[-1].upper() == 'K':
        temp=float(temp)*1000
    else: 
        temp = float(x.strip('€'))
        
    return temp
    
```

```python
data1['Value']=data1['Value'].apply(lambda x:x if x is np.nan else curtonum(x))
# applying function to remove the euro sign and the suffix in value column
```

```python
data1['Wage']=data1['Wage'].apply(lambda x:x if x is np.nan else curtonum(x))
# applying function to remove the euro sign and the suffix in wage column
```

```python
data1['Release Clause']=data1['Release Clause'].apply(lambda x:x if x is np.nan else curtonum(x))
# applying function to remove the euro sign and the suffix in release clause column
```

##### In the case of weight kgs is in the suffix which makes it a string. 'kgs' in the end has to be removed and the column has to be converted into float type to perform analysis

```python
data1['Weight']=data1['Weight'].apply(lambda x:x if x is np.nan else float(x[0:-3]))
# The lamda function removes the last 3 charcters in the string 'kgs' and converts it into float datatype
```

```python
data1['Contract Valid Until'].dtype # to check the datatype of 'Contact Valid Unitil' column
# it is seen that the column is of object datatype
```

```python
data1['Contract Valid Until']=pd.to_datetime(data1['Contract Valid Until']) 
# to convert the column to datetime datatype
```

```python
data1['Joined']=data1['Joined'].apply(lambda x:x if x is np.nan else int(x[-4:]))
# extracting the year joined (last 4 charecters)  from the joined column
```

```python
data1['Height']=data1['Height'].apply(lambda x: x if x is np.nan else float(x.replace("'",'.')))
# replacing the "'" in the data with '.' to convert it into float so analysis can be performed
```

# Null value replacement

```python
data1['Position'].fillna('Unknown',inplace=True)
# replacing the null values in the position with 'Unknown' as it is catagorical and a good estimate of the value cannot
# be made
```

```python
data1['Joined'].fillna(data1['Joined'].median(),inplace=True)
# imputing the missing data in the 'Joined' column with the median as it is not affected by the extreme values
```

```python
data1['Joined']=data1['Joined'].astype(int) # converting the column to int datatype 
```

```python
data1['Contract Valid Until'].fillna(method='bfill',inplace=True)
# filling the contract valid until column using backwards fill which fill the value in the previous row as median is not
# a suitable option for imputing the data value
```

```python
data1['Height'].fillna(data1['Height'].median(),inplace=True)
# imputing the missing data in the 'Height' column with the median as it is not affected by the extreme values
```

```python
data1['Weight'].fillna(data1['Weight'].median(),inplace=True)
# imputing the missing data in the 'Weight' column with the median as it is not affected by the extreme values
```

```python
data1['Release Clause'].fillna(data1['Release Clause'].median(),inplace=True)
# imputing the missing data in the 'Release Clause' column with the median as it is not affected by the extreme values
```

```python
data1['International Reputation'].fillna(data1['International Reputation'].median(),inplace=True)
# imputing the missing data in the 'International Reputation' column with the median as it is not affected by the 
# extreme values
```
